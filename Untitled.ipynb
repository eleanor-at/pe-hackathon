```python
df_france = df[df['Country name'] == 'France']
couples_annee_bonheur = list(zip(df_france['year'], df_france['Positive affect']))
couples_annee_bonheur
```

```python
import matplotlib.pyplot as plt
annees, bonheur = zip(*couples_annee_bonheur)
plt.figure(figsize=(10, 5))
plt.plot(annees, bonheur, marker='o', color='b', linestyle='-', linewidth=2)
plt.title("Évolution du bonheur en France")
plt.xlabel("Année")
plt.ylabel("Score de bonheur")
plt.grid(True)
plt.xticks(annees) 
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Log GDP per capita'], df['Positive affect'], color='b', s=100, alpha=0.6, edgecolors='k')
plt.title("Relation entre la croissance économique et le bonheur")
plt.xlabel("croissance économique")
plt.ylabel("Score de bonheur")
plt.grid(True)
plt.show()
```
